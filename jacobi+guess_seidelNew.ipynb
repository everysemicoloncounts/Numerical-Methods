```python
import numpy as np
from numpy import array,linalg
np.set_printoptions(precision = 8)
np.set_printoptions(suppress = True)

'''
a=np.array([[1,2,3,4],
        [1,2,5,6],
        [6,7,9,3],
        [4,9,1,4]],float)
b=np.array([6,8,9,0],float)
'''
'''
a = np.array([[10,-1,2,0],
             [-1,11,-1,3],
             [2,-1,10,-1],
             [0,3,-1,8]],float)
b = np.array([6,25,-11,15],float)
'''

#a = np.array([[4,3,0],[3,4,-1],[0,-1,4]],float)
#b = np.array([24,30,-24],float)

a=np.array([[-10,-8,0],[-8,10,-1],[0,-1,10]],float)
b=np.array([-6,9,28],float)
w=float(input("required w: "))
n=len(b)
x = np.zeros(n)
xo = np.zeros(n)
TOL=10e-5
k=1

N=int(input("maximum number of iteration: "))
while k<=N:
    for i in range(n):
        x[i] = (1-w)*xo[i] + w*b[i]/a[i,i]
        for j in range(n):
            if j < i:
                x[i] += w*-a[i,j]*x[j]+b[i]/a[i,i]
            if j > i:
                x[i] += w*-a[i,j]*xo[j]+b[i]/a[i,i]
    print(f"Iteration{k}: ",x)
    temp=0
    for i in range(n):
        temp=max(temp,abs(x[i]-xo[i]))
    if temp<TOL:
        break
    k=k+1
    for i in range(n):
        xo[i]=x[i]
if k>N:
    print("maximum number of iterations exceeded", k)
```
